```python
import os
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt

# Set image dimensions
IMG_HEIGHT, IMG_WIDTH = 64, 64  # Updated resolution
CHANNELS = 3  # RGB images

# Load and preprocess dataset
def load_and_preprocess_images(directory_path):
    """Load and preprocess images for training."""
    image_paths = [
        os.path.join(directory_path, fname)
        for fname in os.listdir(directory_path)
        if fname.lower().endswith(('.jpg', '.jpeg', '.png'))
    ]
    dataset = []
    for img_path in image_paths:
        img = tf.io.read_file(img_path)
        img = tf.image.decode_jpeg(img, channels=CHANNELS)
        img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
        img = img / 255.0  # Normalize to [0, 1]
        dataset.append(img)
    return np.array(dataset)

# Variational Autoencoder (VAE) class
class VariationalAutoencoder(Model):
    def __init__(self, latent_dim):
        super(VariationalAutoencoder, self).__init__()
        self.latent_dim = latent_dim

        # Encoder with ResNet-like blocks
        self.encoder = tf.keras.Sequential([
            layers.InputLayer(input_shape=(IMG_HEIGHT, IMG_WIDTH, CHANNELS)),
            layers.Conv2D(64, (4, 4), strides=2, padding='same', activation='relu'),
            layers.Conv2D(128, (4, 4), strides=2, padding='same', activation='relu'),
            layers.Conv2D(256, (4, 4), strides=2, padding='same', activation='relu'),
            layers.Flatten(),
            layers.Dense(512, activation='relu'),
            layers.Dense(latent_dim * 2)  # Mean and log variance
        ])

        # Decoder with Transpose Convolutions
        self.decoder = tf.keras.Sequential([
            layers.InputLayer(input_shape=(latent_dim,)),
            layers.Dense(8 * 8 * 256, activation='relu'),
            layers.Reshape((8, 8, 256)),
            layers.Conv2DTranspose(128, (4, 4), strides=2, padding='same', activation='relu'),
            layers.Conv2DTranspose(64, (4, 4), strides=2, padding='same', activation='relu'),
            layers.Conv2DTranspose(32, (4, 4), strides=2, padding='same', activation='relu'),
            layers.Conv2DTranspose(CHANNELS, (3, 3), padding='same', activation='sigmoid')
        ])

    def reparameterize(self, mean, log_var):
        """Reparameterization trick."""
        eps = tf.random.normal(shape=tf.shape(mean))
        return eps * tf.exp(log_var * 0.5) + mean

    def call(self, x):
        encoded = self.encoder(x)
        mean, log_var = tf.split(encoded, num_or_size_splits=2, axis=1)
        z = self.reparameterize(mean, log_var)
        reconstructed = self.decoder(z)
        return reconstructed, mean, log_var

# Define loss function
def compute_loss(model, x):
    reconstructed, mean, log_var = model(x)
    reconstruction_loss = tf.reduce_mean(
        tf.reduce_sum(tf.keras.losses.binary_crossentropy(x, reconstructed), axis=(1, 2))
    )
    kl_divergence = -0.5 * tf.reduce_sum(1 + log_var - tf.square(mean) - tf.exp(log_var), axis=1)
    kl_divergence = tf.reduce_mean(kl_divergence)
    return reconstruction_loss + kl_divergence

# Training step
@tf.function
def train_step(model, x, optimizer):
    with tf.GradientTape() as tape:
        loss = compute_loss(model, x)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    return loss

```

```python
if __name__ == "__main__":
    data_dir = "/content/data51"  # Path to dataset
    image_dataset = load_and_preprocess_images(data_dir)
    print(f"Dataset Shape: {image_dataset.shape}")

    # Hyperparameters
    latent_dim = 256
    epochs = 25
    batch_size = 64
    learning_rate = 1e-4

    # Prepare the dataset
    dataset = tf.data.Dataset.from_tensor_slices(image_dataset).shuffle(len(image_dataset)).batch(batch_size)

    # Initialize model and optimizer
    vae = VariationalAutoencoder(latent_dim)
    optimizer = tf.keras.optimizers.Adam(learning_rate)

    # Training loop
    for epoch in range(epochs):
        for step, x in enumerate(dataset):
            loss = train_step(vae, x, optimizer)
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss.numpy():.4f}")

    # Generate new images
    def generate_images(model, num_images=5):
        random_latent_vectors = tf.random.normal(shape=(num_images, latent_dim))
        generated_images = model.decoder(random_latent_vectors)
        generated_images = generated_images.numpy()

        plt.figure(figsize=(15, 5))
        for i in range(num_images):
            plt.subplot(1, num_images, i + 1)
            plt.imshow(generated_images[i])
            plt.axis('off')
        plt.show()

    generate_images(vae)
```

```python
# Save the model
vae.encoder.save("car_vae_encoder.h5")
vae.decoder.save("car_vae_decoder.h5")
print("Model saved successfully.")
```

```python
import os
latent_dim = 256

# Load the model
def load_vae_model(encoder_path, decoder_path):
    encoder = tf.keras.models.load_model(encoder_path, compile=False)
    decoder = tf.keras.models.load_model(decoder_path, compile=False)
    return encoder, decoder

# Generate new images from random latent vectors
def generate_images(decoder, num_images, latent_dim):
    random_latent_vectors = tf.random.normal(shape=(num_images, latent_dim))
    generated_images = decoder(random_latent_vectors).numpy()

    # Plot the generated images
    plt.figure(figsize=(7,7))
    for i in range(num_images):
        plt.subplot(int(num_images ** 0.5), int(num_images ** 0.5), i + 1)
        plt.imshow(generated_images[i].squeeze(), cmap="gray")
        plt.axis("off")
    plt.show()

# Example usage
encoder_path = "updated_car_vae_encoder.h5"
decoder_path = "updated_car_vae_decoder.h5"

if os.path.exists(encoder_path) and os.path.exists(decoder_path):
    encoder, decoder = load_vae_model(encoder_path, decoder_path)
    generate_images(decoder, num_images=16, latent_dim=latent_dim)
else:
    print("Model files not found. Train and save the model first.")
```
